```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
import numpy as np
```

```python
from sklearn.datasets.samples_generator import make_blobs
X,Y = make_blobs(n_samples=500,centers=4,cluster_std=0.40,random_state=0)
```

```python
Y
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_predict = kmeans.predict(X)
y_predict
```

```python
plt.scatter(X[:,0],X[:,1],c=y_predict)
# plt.show()
```

```python
center = kmeans.cluster_centers_
center
```

```python
plt.scatter(X[:,0],X[:,1],c = y_predict)
plt.scatter(center[:,0],center[:,1],c='red',s=50)
```

# mean shift algorithm

```python
from sklearn.cluster import MeanShift
```

```python
center =[[2,2],[4,5],[3,10]]
x,_ = make_blobs(n_samples =500,centers=center,cluster_std=1)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
ms = MeanShift()
ms.fit(x)
label = ms.labels_
label
```

```python
cluster_center = ms.cluster_centers_
cluster_center
```

```python
plt.scatter(x[:,0],x[:,1] ,c=label)
plt.scatter(cluster_center[:,0],cluster_center[:,1],c='red')
```

```python
n_cluster = len(np.unique(label))
n= len(cluster_center)
print(n)
print('estimated cluster'+str(n_cluster))
```

```python
colors = 10*['r.','g.','b.','c.','k.','m.']
for i in range(len(x)):
    plt.plot(x[i][0],x[i][1],colors[label[i]],markersize=10)
    plt.scatter(cluster_center[:,0],cluster_center[:,1],marker='X',color='k',s=250,lw=5,zorder=10)
```
