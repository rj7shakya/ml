# Linear Regresstion with Scikit learn

```python
df.head()
```

```python
X = df['RM'].values.reshape(-1,1)
```

```python
y= df['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10));
sns.regplot(X,y);
plt.show()
```

```python
sns.jointplot(x='MEDV',y='RM',data=df,kind='reg',size=10);
plt.show()
```
