```python
from sklearn.cluster import KMeans
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
X= iris["data"]
Y = iris["target"]
feature_names =iris["feature_names"]
feature_names
```

```python
df = pd.DataFrame(X)
df.head()
```

```python
df.columns = feature_names
df.head()
```

```python
df["iris_types"]=Y
df.head()
```

```python
 df.plot(figsize=(11,11))
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,)
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(X)
Y_pred = kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c= Y_pred,cmap=plt.cm.Set1,edgecolors='k')
```

```python
c= kmeans.cluster_centers_
c
```

```python
plt.scatter(X[:,0],X[:,1],c= Y_pred,cmap=plt.cm.Set1,edgecolors='k')
plt.scatter(c[:,0],c[:,1],color='blue',s=50)
```
